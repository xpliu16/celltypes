```python
import pandas as pd
import numpy as np

# From HUGO gene name

df1 = pd.read_csv("HUGO_genes/group-161.csv", header=1)   # Ligand gated ion channels
df2 = pd.read_csv("HUGO_genes/group-168.csv", header=1)   # Amine receptors (non-ionotropic)
df3 = pd.read_csv("HUGO_genes/group-178.csv", header=1)    # Voltage gated ion channels
df4 = pd.read_csv("HUGO_genes/group-1902.csv", header=1)   # Neuropeptides    # This one is missing the group column
```

```python
# Drop Group column since it is missing from df4
df1 = df1.drop(columns=['Group'])
df2 = df2.drop(columns=['Group'])
df3 = df3.drop(columns=['Group'])
```

```python
#d = {'LGIC' : df1, 'AmineR' : df2, 'VGIC': df3, 'NP': df4}
d = {'LGIC' : df1, 'AmineR' : df2, 'VGIC':df3, 'NP':df4}
df = pd.concat(d.values())
```
